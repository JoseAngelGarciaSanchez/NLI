```python
import pandas as pd 
from datasets import load_dataset

snli = load_dataset('snli')
train = pd.DataFrame(snli['train'])
test = pd.DataFrame(snli['test'])
validation = pd.DataFrame(snli['validation'])
```

```python
# Create an instance of the EDA_SNLI class
from class_eda import EDA_SNLI
eda = EDA_SNLI(train, test, validation)
```

```python
#use of the calculate text stats function
eda.calculate_text_stats(train)
eda.calculate_text_stats(test)
eda.calculate_text_stats(validation)
```

```python
#use of the function that maps labels
train_mapped=eda.map_labels(train)
test_mapped=eda.map_labels(test)
validation_mapped=eda.map_labels(validation)
```

```python
#use of the function label distribution
eda.plot_label_distribution(train, 'Label Distribution for Train Database')
eda.plot_label_distribution(test, 'Label Distribution for Test Database')
eda.plot_label_distribution(validation, 'Label Distribution for Validation Database')
```

```python
#use of the function word count
train_premise = eda.word_count(train, 'premise')
train_hypothesis = eda.word_count(train, 'hypothesis')

test_premise = eda.word_count(test, 'premise')
test_hypothesis = eda.word_count(test, 'hypothesis')

validation_premise = eda.word_count(validation, 'premise')
validation_hypothesis = eda.word_count(validation, 'hypothesis')
```

```python
#use of the function to plot sentence length distribution
eda.plot_length_distribution(train_premise, train_hypothesis, 'Sentence length distribution for Train Database')
eda.plot_length_distribution(test_premise, test_hypothesis, 'Sentence length distribution for Test Database')
eda.plot_length_distribution(validation_premise, validation_hypothesis, 'Sentence length distribution for Validation Database')
```
